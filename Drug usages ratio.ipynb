```python
#Convert the features into labels encoded trough a function
from sklearn.preprocessing import LabelEncoder

def label_encode_features(data, columns):
    label_encoder = LabelEncoder()
    for column in columns:
        data[column] = label_encoder.fit_transform(data[column])
    return data

```

```python
# List of columns to encode
columns_to_encode = ['Age', 'Gender', 'Education','Enjoyable with-','Live with','Conflict with law','Motive about drug','Spend most time',
                     'Failure in life','Mental/emotional problem','Suicidal thoughts','Family relationship','Financials of family','Addicted person in family',
                    'no. of friends','Withdrawal symptoms','friends? houses at night','Satisfied with workplace','Case in court','Living with drug user',
                    'Smoking','Ever taken drug','Friends influence','If chance given to taste drugs','Easy to control use of drug','Frequency of drug usage'] 

# Apply label encoding to the specified columns
data_encoded = label_encode_features(data, columns_to_encode)


```
