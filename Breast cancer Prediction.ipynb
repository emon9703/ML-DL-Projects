```python
#Convert the features into labels encoded trough a function
from sklearn.preprocessing import LabelEncoder

def label_encode_features(data, columns):
    label_encoder = LabelEncoder()
    for column in columns:
        data[column] = label_encoder.fit_transform(data[column])
    return data

# List of columns to encode
columns_to_encode = ['Gender','Tumour_Stage','Histology','ER status', 'PR status', 'HER2 status',
       'Patient_Status'] 

# Apply label encoding to the specified columns
data_encoded = label_encode_features(data, columns_to_encode)
```
